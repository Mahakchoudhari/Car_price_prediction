```python
import numpy as np
import pandas as pd

# 🔹 Load data
dataset = pd.read_csv("pricing.csv")
dataset = dataset.drop(columns=["Unnamed: 0"])

# 🔹 Split X, y
X = dataset.drop("selling_price", axis=1)
y = dataset["selling_price"]

# 🔹 Train-Test Split (IMPORTANT: before pipeline fit)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# 🔹 Custom Transformer (FIXED)
from sklearn.base import BaseEstimator, TransformerMixin

class OutlierHandler(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        X = np.array(X)
        self.Q1_ = np.percentile(X, 25, axis=0)
        self.Q3_ = np.percentile(X, 75, axis=0)
        self.IQR_ = self.Q3_ - self.Q1_
        return self
    
    def transform(self, X):
        X = np.array(X)
        lower = self.Q1_ - 1.5 * self.IQR_
        upper = self.Q3_ + 1.5 * self.IQR_
        return np.clip(X, lower, upper)

# 🔹 Preprocessing
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer

num_cols = X.select_dtypes(include=np.number).columns
cat_cols = X.select_dtypes(exclude=np.number).columns

preprocessor = ColumnTransformer([
    ('num', Pipeline([
        ('imputer', SimpleImputer(strategy='mean')),
        ('outlier', OutlierHandler())
    ]), num_cols),

    ('cat', Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('encoder', OneHotEncoder(handle_unknown='ignore'))
    ]), cat_cols)
])

# 🔹 Feature Selection
from sklearn.feature_selection import VarianceThreshold, SelectKBest, f_regression

vt = VarianceThreshold()
selector = SelectKBest(score_func=f_regression, k=10)

# 🔹 Model
from xgboost import XGBRegressor
model = XGBRegressor()

# 🔹 Full Pipeline
pipeline = Pipeline([
    ('preprocess', preprocessor),
    ('variance', vt),
    ('select', selector),
    ('model', model)
])

# 🔥 Train (IMPORTANT)
pipeline.fit(X_train, y_train)

print("Pipeline fitted successfully ✅")

# 🔹 Predict on test
y_pred = pipeline.predict(X_test)

from sklearn.metrics import r2_score
print("R2 Score:", r2_score(y_test, y_pred))

# 🔹 New Input
new_data = pd.DataFrame([{
    'name': 'Maruti',
    'year': 2014,
    'km_driven': 145500,
    'fuel': 'Diesel',
    'seller_type': 'Dealer',
    'transmission': 'Manual',
    'owner': 'First Owner',
    'seats': 5,
    'max_power (in bph)': 74.0,
    'Mileage Unit': 'kmpl',
    'Mileage': 23.4,
    'Engine (CC)': 1248
}])

# 🔥 Predict new data
prediction = pipeline.predict(new_data)

print("Predicted Price:", prediction[0])
```
